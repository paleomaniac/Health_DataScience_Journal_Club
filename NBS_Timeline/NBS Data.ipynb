```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df = pd.read_csv("nchs-infant-and-neonatal-mortality-rates-united-states-1915-2013.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.Type.unique()
```

```python
neo_df = df[df['Type']=='Neonatal']
inf_df = df[df['Type']=='Infant']
```

```python
mortality = sns.lineplot( x= neo_df['Year'], y = neo_df['Mortality Rate'])
plt.axvline(1960,0,0.99, color = 'red', label = '1st PKU Test (1960)')
plt.axvline(1963,0,0.99, color = 'green', label = 'NBS Begins (1963)')
plt.axvline(1966,0,0.99, color = 'grey', ls = '--', label = 'PKU (1966)')
plt.axvline(1968,0,0.99, color = 'grey', ls = ':', label = 'Hyperthroidism (1968)')
plt.axvline(1974,0,0.99, color = 'grey', ls = '-.', label ="Galactosemia (1974)")
plt.axvline(1988,0,0.99, color = 'grey', ls = '-', label ="Sickle Cell + s-Beta thalassemia (1988)")
plt.axvline(1990,0,0.99, color = 'red', label= "Mass Spec (1990)" )
plt.axvline(1992,0,0.99, color = 'grey', ls = '--', label ="Congenital adrenal hyperplasia (1992)")
plt.axvline(2001,0,0.99, color = 'grey', ls = ':', label ="Over 18+ new screening disorders (2001)")
plt.axvline(2006,0,0.99, color = 'grey', ls = '-.', label ="Cystic Fibrosis (2006)")
plt.legend(loc = 'upper center',bbox_to_anchor = (1.4, 0.9), title = "NBS Events")
plt.title("Neonatal Mortality and Screening Timeline")
plt.savefig("NBS_Neo_Mortality.png", bbox_inches='tight')
```
